# Let's Findout he Gender Distribution According to the Target Value

```python
sns.countplot(x='sex',hue = "target",data = data)
plt.xticks([0,1],["Female","Male"])
plt.legend(labels = ['No-Disease','Disease'])
plt.show()
```

```python
sns.distplot(data['age'],bins = 20)
plt.show()
```

From this plot we can say that most of the people having heart disease had aged between 50 to 60

# FInding out Which Chest pain type is more common between people

```python
sns.countplot(data['cp'])
plt.xticks([0,1,2,3],['typical angina','atypical angin','non-anginal pain','asymptomatic'])
plt.xticks(rotation = 45)
plt.xlabel('Chest Pain Types')
plt.show()
```

From this plot we can say that Chest pain type 0 i.e, Typical angina is more common between people

```python
sns.countplot(x = 'cp', hue ='target',data = data)
plt.legend(labels =["No-Disease","Disease"])
plt.show()
```

From this graph we can say that even healthy people i.e, cp type non-anginal pain(value = 2) are having heart disease. Since Chest pain is subjective due to stress, Physical activities, etc. And it varies between gender.

# Let's show FBS(Fasting Blood Sugar) Distribution according to Target value

```python
sns.countplot(x = 'fbs', hue ='target',data = data)
plt.legend(labels =["No-Disease","Disease"])
plt.show()
```

If FBS greater than 120 we can say that the person having diabetes. From this plot we can say that, More number of Heart disease patients did not have diabetes.

# Let's Check the Resting Blood Pressure Distribution

# Comparing trestbps(Resting Blood Pressure) as per Sex Column.

```python
g = sns.FacetGrid(data,hue = "sex",aspect = 4)
g.map(sns.kdeplot,'trestbps',shade = True)
plt.legend(labels = ['Male', ' Female'])
```

From this plot we can see that Female have Low Resting Blood Pressure as compared to Male.
For Female it is around 120 while for Male it is little less than 140.
